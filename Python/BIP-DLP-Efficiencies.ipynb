Formato modelo primal

```python
instance_name = "250-10-10-1-4"
filename = "EfficiencyMeasures/"+instance_name+".txt"
i1 = open(filename, "r")

data = i1.read().split('\n')
n, m, q = data[0].split(' ')
n = int(n)
m = int(m)
q = int(q)
b1 = data[2+m].split(' ')
b2 = data[3+m+q].split(' ')

c = data[1].split(' ')[:-1]
for j in range(n):
    c[j] = c[j]+"*x"+str(j+1)
c = ' + '.join(c)
print("max: "+c+";\n")

for i in range(2,2+m):
    a = data[i].split(' ')[:-1]
    for j in range(n):
        a[j] = a[j]+"*x"+str(j+1)
    a = ' + '.join(a)
    print(a+" <= "+b1[i-2]+";")

print("")
for i in range(3+m,3+m+q):
    a = data[i].split(' ')[:-1]
    for j in range(n):
        a[j] = a[j]+"*x"+str(j+1)
    a = ' + '.join(a)
    print(a+" >= "+b2[i-3-m]+";")

print("")
for i in range(n):
    print("x"+str(i+1)+" <= 1;")
    
i1.close()
```

Formato para modelo dual

```python
i2 = open(filename, "r")
data2 = i2.read().split('\n')
n, m, q = data2[0].split(' ')
n = int(n)
m = int(m)
q = int(q)
b1 = data2[2+m].split(' ')[:-1]
b2 = data2[3+m+q].split(' ')[:-1]

c = data2[1].split(' ')[:-1]
for i in range(m):
    b1[i] = b1[i]+"*u"+str(i+1)
b1 = ' + '.join(b1)

for i in range(q):
    b2[i] = b2[i]+"*u"+str(m+i+1)
b2 = ' - '.join(b2)

v = []
for j in range(n):
    v.append("v"+str(j+1))
v = ' + '.join(v)

print("min: "+b1+" - "+b2+" + "+v+";\n")

A = np.zeros((m+q,n),type(str))

for i in range(2,2+m):
    A[i-2] = data2[i].split(' ')[:-1]
for i in range(3+m,3+m+q):
    A[i-3] = data2[i].split(' ')[:-1]

for j in range(n):
    a1 = []
    for i in range(m):
        a1.append(A[i][j]+"*u"+str(i+1))
    a2 = []
    for i in range(q):
        a2.append(A[m+i][j]+"*u"+str(m+i+1))
    a1 = ' + '.join(a1)
    a2 = ' - '.join(a2)
    print(a1+' - '+a2+' + v'+str(j+1)+' >= '+c[j]+';')
    
i2.close()
```
